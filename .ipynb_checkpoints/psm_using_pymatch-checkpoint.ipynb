```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from pymatch.Matcher import Matcher
```

#### Read CSV File

```python
df = pd.read_csv('Arthritis_on_heart_attack.csv', index_col='Unnamed: 0')
df.head()
```

```python
df = df.rename(columns={'arthritis.type': 'arthritis', 'physical.activity': 'physical', 'medical.access':'medical','blood.pressure':'bp', 'healthy.diet':'diet',
                       'covered.health' : 'chealth', 'heart.attack':'attack'})
df.head()
```

#### We have to convert categorical variables into numeric variables

```python
labelencoder = LabelEncoder()
for col in df.columns:
    df[col] = labelencoder.fit_transform(df[col])
df.head()
```

```python
for col in df.columns:
    df[col] = df[col].astype('category')
df.dtypes
```

#### Create test and control groups. It will be used in the logistic regression model used to generate propensity scores.

```python
test = df[df.arthritis == 1]
control = df[df.arthritis == 0]
control.head()
```

#### Matcher object is initialized and model is fitted

```python
m = Matcher(test, control, yvar='arthritis' ,exclude=['arthritis','attack'])
```

```python
np.random.seed(20190101)
m.fit_scores(balance=True, nmodels=30)
```

```python
m.predict_scores()
```

#### Visualizing propensity score distribution before matching between treatment and control groups

```python
m.plot_scores()
```

```python
m.match(method="min", nmatches=1, threshold=0.2)
m.record_frequency()
```

```python
df_match = m.matched_data.sort_values("match_id")
df_match.head(6)
```

#### Data frame consisting of  matched data. Matcher has also assigned a match_id corresponding to matched record for both treatment and control group

```python
m_treated = df_match[df_match.arthritis ==1]
m_treated_propensity = m_treated['scores']
m_control = df_match[df_match.arthritis ==0]
m_control_propensity = m_control['scores']
```

#### Visualizing propensity score distriution before matching between treatment and control groups

```python
plt.rcParams['figure.figsize'] = 8, 6
plt.rcParams['font.size'] = 12

sns.distplot(m_control_propensity, label='control')
sns.distplot(m_treated_propensity, label='treatment')
plt.xlim(0, 1)
plt.title('Propensity Score Distribution of Control vs Treatment after matching')
plt.ylabel('Density')
plt.xlabel('Scores')
plt.legend()
plt.show()
```

#### Determine if the data is balanced across the covariates to detect any statistical differences between the covariates of our matched test and control groups

```python
categorical_results = m.compare_categorical(return_table=True)
```

```python
categorical_results
```

#### Looking at the plots and test results, there is a nice balance in the categorical features. The p-values from the Chi-Square tests are all > 0.05 and therefore there is not enough evidence to conclude that the variables are associated.


#### In the problem, the following hypothesis was assumed:
#### $H_0$ - Rheumatoid arthritis patients are not prone to heart attack
#### $H_a$ - Rheumatoid arthritis patients are prone to heart attack
####  Using a paired t-test,  the p-value was found to be < 0.05, therefore, the null hypothesis can be rejected and alternative hypothesis can be accepted.
#### From this analysis, it is found that patients with Rheumatoid Arthritis (RA) are more likely to get a heart-attack.

```python
stats.ttest_rel(m_treated['attack'].values, m_control['attack'].values)
```
